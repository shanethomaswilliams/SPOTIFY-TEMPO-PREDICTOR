## Proof Of Concept ML Problem

```python
from sklearn.neural_network import MLPRegressor
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

```python
## Getting our data
artists = pd.read_csv("ml_artists.csv")
track = pd.read_csv("ml_track.csv")
display(artists)
display(track)
```

```python
## Fixing null values to 0, as they indicate 0 awards and scaling our data
artists = artists.replace(np.nan, 0)
display(artists)

X_unscaled = artists.iloc[:, 2:]
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X_unscaled)
display(X_unscaled)
display(X_scaled)

display(track.iloc[:, 1:])
```

```python
X = X_scaled
y = (track.iloc[:, 1:]).to_numpy()
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y)
model = MLPRegressor(random_state=1, max_iter=500).fit(Xtrain, ytrain)

print(model.predict(Xtest))
print(ytest)

results = pd.DataFrame([model.predict(Xtest), ytest]).T
display(results)
print("MODEL SCORE:")
print(model.score(Xtest, ytest))
```
